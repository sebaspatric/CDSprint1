###Ejercicio 1:
1. Diseñe e implemente una clase Producto con los siguientes atributos: id, nombre, descripción, cantidad y precio. La clase debe incluir métodos para obtener y modificar la información de los productos.

```python
class Producto:
    def __init__(self, id, nombre, descripcion, cantidad, precio):
        self.id = id
        self.nombre = nombre
        self.descripcion = descripcion
        self.cantidad = cantidad
        self.precio = precio

    def obtener_id(self):
        return self.id

    def obtener_nombre(self):
        return self.nombre

    def obtener_descripcion(self):
        return self.descripcion

    def obtener_cantidad(self):
        return self.cantidad

    def obtener_precio(self):
        return self.precio

    def modificar_id(self, id):
        self.id = id

    def modificar_nombre(self, nombre):
        self.nombre = nombre

    def modificar_descripcion(self, descripcion):
        self.descripcion = descripcion

    def modificar_cantidad(self, cantidad):
        try:
            cantidad = float(cantidad)
            self.cantidad = cantidad
        except ValueError:
            print("Error: La cantidad debe ser un valor numérico.")

    def modificar_precio(self, precio):
        try:
            precio = float(precio)
            self.precio = precio
        except ValueError:
            print("Error: El precio debe ser un valor numérico.")

# Ejemplo de uso
if __name__ == "__main__":
    # Crear un objeto Producto
    producto1 = Producto(id=1, nombre="Producto1", descripcion="Descripción del Producto1", cantidad=10, precio=100)

    # Obtener información del producto
    print("ID:", producto1.obtener_id())
    print("Nombre:", producto1.obtener_nombre())
    print("Descripción:", producto1.obtener_descripcion())
    print("Cantidad:", producto1.obtener_cantidad())
    print("Precio:", producto1.obtener_precio())

    # Modificar información del producto
    producto1.modificar_nombre("Nuevo Nombre")
    producto1.modificar_cantidad(20)

    # Obtener información actualizada del producto
    print("\nInformación actualizada del producto:")
    print("ID:", producto1.obtener_id())
    print("Nombre:", producto1.obtener_nombre())
    print("Descripción:", producto1.obtener_descripcion())
    print("Cantidad:", producto1.obtener_cantidad())
    print("Precio:", producto1.obtener_precio())

```

###Ejercicio 2:
2. Diseñe e implemente una clase Inventario que utilice una estructura de datos (por ejemplo, un diccionario) para almacenar objetos de la clase Producto. La clase Inventario debe incluir métodos para realizar las siguientes operaciones:
* Agregar un producto al inventario
* Eliminar un producto del inventario
* Actualizar la información de un producto en el inventario
* Buscar un producto en el inventario por su id
* Listar todos los productos en el inventario, incluyendo información como la cantidad total de productos y el valor total del inventario

```python
class Inventario:
    def __init__(self):
        self.productos = {}

    def agregar_producto(self, producto):
        try:
            id = int(producto.obtener_id())
            cantidad = int(producto.obtener_cantidad())
            precio = float(producto.obtener_precio())
            self.productos[id] = producto
            print("Producto agregado al inventario.")
        except ValueError:
            print("Error: Los datos del producto deben ser numéricos.")

    def eliminar_producto(self, id):
        try:
            del self.productos[id]
            print(f"Producto con ID {id} eliminado del inventario.")
        except KeyError:
            print("No se encontró ningún producto con ese ID en el inventario.")

    def actualizar_producto(self, id, nuevo_producto):
        try:
            id = int(nuevo_producto.obtener_id())
            cantidad = int(nuevo_producto.obtener_cantidad())
            precio = float(nuevo_producto.obtener_precio())
            self.productos[id] = nuevo_producto
            print(f"Información del producto con ID {id} actualizada.")
        except ValueError:
            print("Error: Los datos del producto deben ser numéricos.")
        except KeyError:
            print("No se encontró ningún producto con ese ID en el inventario.")

    def buscar_producto_por_id(self, id):
        try:
            return self.productos[id]
        except KeyError:
            print("No se encontró ningún producto con ese ID en el inventario.")
            return None

    def listar_productos(self):
        total_productos = 0
        valor_total_inventario = 0

        print("Productos en el inventario:")
        for producto in self.productos.values():
            print(f"ID: {producto.obtener_id()}, Nombre: {producto.obtener_nombre()}, Cantidad: {producto.obtener_cantidad()}, Precio: {producto.obtener_precio()}")
            total_productos += float(producto.obtener_cantidad())
            valor_total_inventario += float(producto.obtener_cantidad()) * float(producto.obtener_precio())


        print(f"Total de productos: {total_productos}")
        print(f"Valor total del inventario: {valor_total_inventario}")

def main():
    """
    Función principal del programa.
    Crea un inventario, agrega algunos productos y muestra la lista de productos en el inventario.
    """
    
    inventario = Inventario()

    # Crear productos
    producto1 = Producto(1, "Producto1", "Descripción1", 10, 100)
    producto2 = Producto(2, "Producto2", "Descripción2", 20, 200)
    producto3 = Producto(3, "Producto3", "Descripción3", 30, 300)
   
    # Agregar productos al inventario
    inventario.agregar_producto(producto1)
    inventario.agregar_producto(producto2)
    inventario.agregar_producto(producto3)
   
    # Listar productos
    inventario.listar_productos()
   
    # Actualizar producto
    producto_actualizado = Producto(2, "Producto2 Modificado", "Descripción2 Modificada", 25, 250)
    inventario.actualizar_producto(2, producto_actualizado)
   
    # Listar productos actualizados
    inventario.listar_productos()
   
    # Eliminar producto
    inventario.eliminar_producto(3)
   
    # Listar productos después de eliminar
    inventario.listar_productos()
   

# Ejemplo de uso
if __name__ == "__main__":
    main()

```

###Ejercicio 3:
3. Implemente una menú de usuario que permita a los usuarios interactuar con el sistema de gestión de inventario. El menú de incluir opciones para realizar las diferentes operaciones disponibles en la clase Inventario. Utilice un bucle while y estructuras de control de flujo if-elif-else para manejar las diferentes opciones del menú. (opcional)

```python
import os
class MenuInventario:
    def __init__(self):
        self.inventario = Inventario()

    def limpiar_consola(self):
        # Para Windows
        if os.name == 'nt':
            _ = os.system('cls')
        # Para otros sistemas (Linux/Mac)
        else:
            _ = os.system('clear')

    def mostrar_menu(self):
        print("\nMenú de Gestión de Inventario:")
        print("1. Agregar producto")
        print("2. Eliminar producto")
        print("3. Actualizar producto")
        print("4. Buscar producto por ID")
        print("5. Listar todos los productos")
        print("6. Salir")

    def ejecutar_opcion(self, opcion):
        if opcion == "1":
            self.agregar_producto()
        elif opcion == "2":
            self.eliminar_producto()
        elif opcion == "3":
            self.actualizar_producto()
        elif opcion == "4":
            self.buscar_producto_por_id()
        elif opcion == "5":
            self.listar_productos()
        elif opcion == "6":
            print("Saliendo del sistema de gestión de inventario.")
        else:
            print("Opción no válida. Por favor, seleccione una opción válida.")

    def agregar_producto(self):
        id = input("Ingrese el ID del producto: ")
        nombre = input("Ingrese el nombre del producto: ")
        descripcion = input("Ingrese la descripción del producto: ")
        cantidad = input("Ingrese la cantidad del producto: ")
        precio = input("Ingrese el precio del producto: ")

        producto = Producto(id, nombre, descripcion, cantidad, precio)
        self.inventario.agregar_producto(producto)

    def eliminar_producto(self):
        id = input("Ingrese el ID del producto a eliminar: ")
        self.inventario.eliminar_producto(id)

    def actualizar_producto(self):
        id = input("Ingrese el ID del producto a actualizar: ")
        nuevo_id = input("Ingrese el nuevo ID del producto: ")
        nombre = input("Ingrese el nuevo nombre del producto: ")
        descripcion = input("Ingrese la nueva descripción del producto: ")
        cantidad = input("Ingrese la nueva cantidad del producto: ")
        precio = input("Ingrese el nuevo precio del producto: ")

        nuevo_producto = Producto(nuevo_id, nombre, descripcion, cantidad, precio)
        self.inventario.actualizar_producto(id, nuevo_producto)

    def buscar_producto_por_id(self):
        id = input("Ingrese el ID del producto a buscar: ")
        producto = self.inventario.buscar_producto_por_id(id)
        if producto:
            print("Producto encontrado:")
            print(f"ID: {producto.obtener_id()}, Nombre: {producto.obtener_nombre()}, Descripción: {producto.obtener_descripcion()}, Cantidad: {producto.obtener_cantidad()}, Precio: {producto.obtener_precio()}")
        else:
            print("Producto no encontrado.")

    def listar_productos(self):
        self.inventario.listar_productos()

    def presionar_tecla_para_continuar(self):
        input("Presione una tecla para continuar...")

    def iniciar_menu(self):
        while True:
            self.limpiar_consola()
            self.mostrar_menu()
            opcion = input("Seleccione una opción: ")
            self.ejecutar_opcion(opcion)
            if opcion == "6":
                break
            self.presionar_tecla_para_continuar()  # Pausar la ejecución antes de continuar

# Ejemplo de uso
if __name__ == "__main__":
    menu = MenuInventario()
    menu.iniciar_menu()

```
